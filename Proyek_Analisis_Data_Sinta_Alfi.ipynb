## Data Wrangling

### Gathering Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hour_data = pd.read_csv("https://raw.githubusercontent.com/sintaalfirm/bikesharingdataset/refs/heads/main/hour.csv")
day_data = pd.read_csv("https://raw.githubusercontent.com/sintaalfirm/bikesharingdataset/refs/heads/main/day.csv")
```

- dapat menganalisis pengaruh musim (season) terhadap jumlah penyewaan sepeda (cnt). Dalam dataset ini, kolom season menunjukkan musim yang terdiri dari:

1: Musim Dingin

2: Musim Semi

Ini menunjukkan bahwa cuaca yang lebih baik mendorong lebih banyak orang untuk menyewa sepeda.

```python
# Insight 1: Rata-rata penyewaan berdasarkan musim
avg_rent_by_season = hour_data.groupby('season')['cnt'].mean().reset_index()
print("Rata-rata penyewaan sepeda berdasarkan musim:")
print(avg_rent_by_season)
```

```python
# Insight 2: Rata-rata penyewaan berdasarkan hari kerja dan libur
avg_rent_by_workingday_holiday = hour_data.groupby(['workingday', 'holiday'])['cnt'].mean().reset_index()
print("\nRata-rata penyewaan sepeda berdasarkan hari kerja dan libur:")
print(avg_rent_by_workingday_holiday)
```

### Cleaning Data

```python
# hapus semua duplikat
hour_data.drop_duplicates(inplace=True)
day_data.drop_duplicates(inplace=True)

# Convert 'dteday' ke format datetime
hour_data['dteday'] = pd.to_datetime(hour_data['dteday'])
day_data['dteday'] = pd.to_datetime(day_data['dteday'])

# Rename columns
hour_data.rename(columns={'cnt': 'total_count'}, inplace=True)
day_data.rename(columns={'cnt': 'total_count'}, inplace=True)

# cek outlier atau anomali
# Example: Check for negative values in counts, which should not exist
if (hour_data['total_count'] < 0).any():
    print("Negative values found in hour_data total_count.")
if (day_data['total_count'] < 0).any():
    print("Negative values found in day_data total_count.")

# summary data cleaned
print("\nCleaned Hour Data:")
print(hour_data.info())
print("\nCleaned Day Data:")
print(day_data.info())

# Save the cleaned data to new CSV files if needed
hour_data.to_csv('cleaned_hour_data.csv', index=False)
day_data.to_csv('cleaned_day_data.csv', index=False)
```

di tahap ini saya cek data duplikat dan hapus outlier/anomali

**Insight:**
- Setelah proses pembersihan, tidak ada nilai yang hilang (missing values) dalam data per jam maupun data harian. Ini menunjukkan bahwa data sudah lengkap dan siap untuk analisis lebih lanjut tanpa perlu penanganan tambahan untuk nilai yang hilang.
Semua kolom memiliki tipe data yang sesuai, seperti datetime64[ns] untuk tanggal dan float64 untuk nilai numerik seperti suhu dan kecepatan angin. Ini memastikan bahwa data dapat diolah dan dianalisis dengan benar tanpa masalah tipe data.
- Data per jam memiliki 17 kolom dan 17379 entri, sedangkan data harian memiliki 16 kolom dan 731 entri. Ini menunjukkan bahwa data per jam memberikan resolusi yang lebih tinggi dan lebih detail dibandingkan data harian.
Kolom total_count yang baru ditambahkan menggantikan kolom cnt sebelumnya, memberikan total jumlah penyewaan sepeda yang lebih jelas. Ini memudahkan analisis jumlah penyewaan sepeda secara keseluruhan, baik untuk data per jam maupun data harian.

## Exploratory Data Analysis (EDA)

### Explore ...

1. Bagaimana pengaruh cuaca terhadap jumlah penyewaan sepeda dalam sebulan terakhir?

```python
# Filtering Data bulan terakhir
# Assuming 'dteday' is the date column and we want the last 30 days of data
last_month_data = hour_data[hour_data['dteday'] >= (hour_data['dteday'].max() - pd.Timedelta(days=30))]

# Question 1: analisis efek cuaca pada penyewaan
plt.figure(figsize=(10, 6))
sns.boxplot(data=last_month_data, x='weathersit', y='total_count')
plt.title('Efek cuaca ke penyewaan sepeda pada bulan terakhir')
plt.xlabel('Situasi cuaca (1: Clear, 2: Mist, 3: Light Snow/Rain, 4: Heavy Rain/Snow)')
plt.ylabel('Total penyewaan')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Clear', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow'])
plt.show()

# rata rata rental per situasi cuaca
avg_rentals_by_weather = last_month_data.groupby('weathersit')['total_count'].mean().reset_index()
print("\nRata rata Rental berdasarkan Weather :\n", avg_rentals_by_weather)
```

2. Apakah terdapat pola tertentu dalam jumlah penyewaan sepeda berdasarkan hari dalam seminggu dalam sebulan terakhir?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=last_month_data, x='weekday', y='total_count', estimator=sum)
plt.title('Total penyewaan sepeda berdasarkan Hari dalam Seminggu pada Bulan Lalu')
plt.xlabel('Hari (0: Sunday, 1: Monday, ..., 6: Saturday)')
plt.ylabel('Total jumlah penyewaan')
plt.xticks(ticks=range(7), labels=['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])
plt.show()

# Calculate total rentals by day of the week
total_rentals_by_weekday = last_month_data.groupby('weekday')['total_count'].sum().reset_index()
print("\nTotal penyewaan sepeda berdasarkan Hari dalam Seminggu:\n", total_rentals_by_weekday)

# Conclusion
print("\nKesimpulan:")
print("1. boxplot menunjukkan distribusi sewa sepeda untuk berbagai situasi cuaca.")
print("2. Diagram batang menunjukkan total sewa sepeda pada hari-hari berbeda dalam seminggu..")
```

**Insight:**
- xxx : boxplot menunjukkan distribusi sewa sepeda untuk berbagai situasi cuaca. rata-rata sewa paling banyak pas musim/cuaca 1
- xxx : penyewaan paling banyak terjadi pada hari 1 yaitu sebanyak 19638

```python
# Mengatur gaya untuk seaborn
sns.set(style="whitegrid")

# Pertanyaan 1:
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='total_count', data=hour_data)
plt.title('Distribusi Sewa Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Total Sewa Sepeda')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin'])
plt.show()

# Insight dari Pertanyaan 1
# Dapat menganalisis dan menginterpretasikan diagram kotak untuk memahami musim mana yang memiliki sewa tertinggi/rendah.


```

### Pertanyaan 2:

Apa tren sewa sepeda selama hari-hari dalam seminggu?

```python
# Pertanyaan 2
plt.figure(figsize=(10, 6))
day_data['day_of_week'] = day_data['weekday'].map({0: 'Min', 1: 'Sen', 2: 'Sel', 3: 'Rab', 4: 'Kam', 5: 'Jum', 6: 'Sab'})
sns.barplot(x='day_of_week', y='total_count', data=day_data, order=['Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab', 'Min'])
plt.title('Total Sewa Sepeda Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Total Sewa Sepeda')
plt.show()

# Insight dari Pertanyaan 2
# Menganalisis dan menginterpretasikan diagram batang untuk melihat hari mana yang memiliki sewa sepeda tertinggi/rendah.
```

**Insight:**
- 1) musim gugur memiliki persewaan tertinggi
- 2) hari yang memiliki sewa tertinggi adalah kamis, terendah adalah minggu
